```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Load Dataset
df = pd.read_csv("Titanic-Dataset.csv")

# Basic Info
print(" Dataset Overview:")
df.info()

print("\n Summary Statistics:")
print(df.describe(include='all'))
```

```python
print("\n Mean = \n", df.mean(numeric_only=True))
print("\n Median = \n", df.median(numeric_only=True))
print("\n Standard Deviation = \n", df.std(numeric_only=True))
```

```python
df.hist(figsize=(12, 8), bins=30, edgecolor='black', color='skyblue')
plt.suptitle(" Histograms of Numerical Features")
plt.tight_layout()
plt.show()
```

```python
numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch']

for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(y=df[col], color='lightcoral')
    plt.title(f" Boxplot of {col}")
    plt.grid(True)
    plt.show()
```

```python
sns.pairplot(df[['Survived', 'Age', 'Fare', 'SibSp', 'Parch']], hue='Survived')
plt.suptitle(" Pairplot by Survival", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title(" Correlation Matrix")
plt.show()
```
